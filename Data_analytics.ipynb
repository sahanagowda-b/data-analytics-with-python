## Matplotlib is a powerfull tool for data visualization

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,8,27,64,125]

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basis")
plt.show()
```

```python
x1 = [1,2,3,4,5]
y2 = [1,8,27,64,125]
plt.plot(x1,y2)
plt.xlabel('Marks')
plt.ylabel('Grade')
plt.title('Distribution')

plt.plot(x,y,color="red",marker="*",linewidth=3,markersize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,5))

x1=[1,2,3,4,5]
x2=[6,7,8,9,10]
x3=[2,4,6,8,10]
x4=[1,3,5,7,9]

plt.subplot(2,2,1)
plt.plot(x1,x2)
plt.title("1st Plot")

plt.subplot(2,2,2)
plt.plot(x2,x3)
plt.title("2nd Graph")

plt.subplot(2,2,3)
plt.plot(x1,x3)
plt.title("3nd Graph")

plt.subplot(2,2,4)
plt.plot(x3,x4)
plt.title("4nd Graph")

plt.show()
```

```python
import matplotlib.pyplot as plt

#bar plot
categories = ["a","b","c","d","e"]
values = [10,20,30,40,50]

plt.bar(categories,values,color='green')
```

```python
#Histogram
data = [1,1,1,2,2,3,3,4,5,6,7,7,7,7,8,8,9,10,10]
plt.hist(data,bins=5)
```

```python
#scatter plot

x1 = [1,2,3,4,5]
y1 = [6,7,8,9,10]

plt.scatter(x1,y1,color='pink')
```

```python
#pie chart

labels = ['a','b','c','d','e']
sizes = [20,25,45,10,5]
colours = ['gold','pink','green','black','brown']
explode=(0,1,0,0,0)
plt.pie(sizes,labels=labels,colors=colours,explode=explode)
```

## seaborn - a powerfull data visualization tool powerfull than matplotlib

```python
!pip install seaborn
```

```python
import seaborn as sb
```

```python
# in-built database
df = sb.load_dataset('tips')
df.head()
```

```python
sb.scatterplot(x='total_bill',y='tip',data=df)
plt.title("Scatter Plot")
plt.show()
```

```python
#line plot

sb.lineplot(x="size",y="total_bill",data=df)
plt.title("Line plot")
plt.show()
```

```python
sb.barplot(x="day",y="total_bill",data=df)
plt.title("Bar-plot")
plt.show()
```

```python
sb.boxplot(x="day",y="total_bill",data=df)
```

```python
#histogram

df.head()
```

```python
sb.histplot(df['total_bill'],bins=10,kde=True)
```

```python
sb.pairplot(df)
```

```python
#kde plot
sb.kdeplot(df['total_bill'],shade=True)
```

```python
#Co-relation btwn variables
corr = df[['total_bill','tip','size']].corr()
sb.heatmap(corr,annot=True,cmap='coolwarm')
```
